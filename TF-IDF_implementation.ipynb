   <h4><strong>What does tf-idf mean?</strong></h4>

   <p>    
Tf-idf stands for <em>term frequency-inverse document frequency</em>, and the tf-idf weight is a weight often used in information retrieval and text mining. This weight is a statistical measure used to evaluate how important a word is to a document in a collection or corpus. The importance increases proportionally to the number of times a word appears in the document but is offset by the frequency of the word in the corpus. Variations of the tf-idf weighting scheme are often used by search engines as a central tool in scoring and ranking a document's relevance given a user query.
</p>
    
   <p>
One of the simplest ranking functions is computed by summing the tf-idf for each query term; many more sophisticated ranking functions are variants of this simple model.
</p>
    
   <p>
Tf-idf can be successfully used for stop-words filtering in various subject fields including text summarization and classification.
</p>
    
</font>

<font face='georgia'>
    <h4><strong>How to Compute:</strong></h4>

Typically, the tf-idf weight is composed by two terms: the first computes the normalized Term Frequency (TF), aka. the number of times a word appears in a document, divided by the total number of words in that document; the second term is the Inverse Document Frequency (IDF), computed as the logarithm of the number of the documents in the corpus divided by the number of documents where the specific term appears.

 <ul>
    <li>
<strong>TF:</strong> Term Frequency, which measures how frequently a term occurs in a document. Since every document is different in length, it is possible that a term would appear much more times in long documents than shorter ones. Thus, the term frequency is often divided by the document length (aka. the total number of terms in the document) as a way of normalization: <br>

$TF(t) = \frac{\text{Number of times term t appears in a document}}{\text{Total number of terms in the document}}.$
</li>
<li>
<strong>IDF:</strong> Inverse Document Frequency, which measures how important a term is. While computing TF, all terms are considered equally important. However it is known that certain terms, such as "is", "of", and "that", may appear a lot of times but have little importance. Thus we need to weigh down the frequent terms while scale up the rare ones, by computing the following: <br>

$IDF(t) = \log_{e}\frac{\text{Total  number of documents}} {\text{Number of documents with term t in it}}.$
for numerical stabiltiy we will be changing this formula little bit
$IDF(t) = \log_{e}\frac{\text{Total  number of documents}} {\text{Number of documents with term t in it}+1}.$
</li>
</ul>

<br>
<h4><strong>Example</strong></h4>
<p>

Consider a document containing 100 words wherein the word cat appears 3 times. The term frequency (i.e., tf) for cat is then (3 / 100) = 0.03. Now, assume we have 10 million documents and the word cat appears in one thousand of these. Then, the inverse document frequency (i.e., idf) is calculated as log(10,000,000 / 1,000) = 4. Thus, the Tf-idf weight is the product of these quantities: 0.03 * 4 = 0.12.
</p>

### Your custom implementation

```python
# You are not supposed to use any other library apart from the ones given below

from collections import Counter
from tqdm import tqdm
from scipy.sparse import csr_matrix
import math
import operator
from sklearn.preprocessing import normalize
import numpy
```

#### fit method implementation
The **fit()** method will return the vocabulary in alphabatically sorted order

**input** ▶ dataset/corpus that should be list of documents

**output** ▶ <font color = "yellow">{word:dimension_number}</font> pair of each word in alphabatical sorted order


```python
# it accepts only list of sentances
def fit(dataset):
    unique_words = set() # at first we will initialize an empty set
    # check if its list type or not
    if isinstance(dataset, (list,)):
        for row in dataset: # for each review in the dataset
            for word in row.split(" "): # for each word in the row/review/document.
                #if the word has only one character then that can be punctuation or special symbol
                #so don't add this into the vocab of unique words
                if len(word) < 2:
                    continue
                unique_words.add(word)
        #sort the unique words alphabatically
        unique_words = sorted(list(unique_words))
        #make the dictionary of the unique words where the key is the words and the value is the dimension_number of the word
        vocab = {words:dimension_number for dimension_number,words in enumerate(unique_words)}
        return vocab
    else:
        print("you need to pass list of sentance")
```

```python
vocab = fit(corpus)

#print alphabatically sorted vocab
print(vocab)

#compare the custom implemented fit() method returned vocab keys(features) with sklearn returned features of the corpus
print(list(vocab.keys()) == vectorizer.get_feature_names())
```

##### IDF(inverse-document-frequency) implementation

```python
#Inverse Document Frequency(IDF) calculation of each words of vocab
#There are 9 words in the vocab, so we calculate IDF for each of these 9 words
def IDF(dataset, vocab):
  #dictionary of words:idf
  idf_values = {}
  print(type(idf_values))
  for words in vocab:
    no_of_documents_with_term_t = 0
    for row in dataset:
      #calculate no_of_documents_with_term_t in the corpus
      if words in row: no_of_documents_with_term_t += 1
    #calculate the idf of each words/terms and append into the idf_values
    idf_values[words] = 1 + math.log(((1+len(dataset))/(1+no_of_documents_with_term_t)), math.e)   #ln5 = math.log(5, math.e)
  return idf_values
```

```python
#compare custom implemented idf with sklearn idf
inverse_document_frequency = IDF(corpus, vocab)
print(list(inverse_document_frequency.values()) == vectorizer.idf_)
```

```python
#implementation of "transform()" method
def transform(dataset, vocab):
  #create 'd' type of empty sparse matrix of size MxN  ==>(M=total_no_of_documents, N=len(vocab))
  sparse_tfidf_matrix = csr_matrix((len(dataset), len(vocab)), dtype='d')
  #take each row of the document
  for row_index in range(len(dataset)):
    word_len_gt2 = [wd for wd in dataset[row_index].split(" ") if len(wd) >= 2 ]
    freq = dict(Counter(word_len_gt2))
    #take each feature/key of the vocab
    for feature, column_index in vocab.items():
      #calculate tfidf
      if (feature in freq):
        tfidf = (freq[feature]/len(word_len_gt2)) * inverse_document_frequency[feature]   #tfidf = tf * idf, tf = (freq[feature]/len(word_len_gt2))
        #insert tfidf into sparse matrix
        sparse_tfidf_matrix[row_index, column_index] = tfidf

  #do L2 normalization on sparse matrix
  output_matrix = normalize(sparse_tfidf_matrix, norm='l2', axis = 1, copy=True, return_norm=False)
  return output_matrix
```

```python
custom_output = transform(corpus, vocab)
print(f"output of custom implemented TFIDF::\n{custom_output}")
print(f"\noutput of sklearn implemented TFIDF::\n{skl_output}")
```

```python
#comparision b/w sklearn implemented and custom implemented tfidf.
print(f"output of custom implemented TFIDF::\n{custom_output.todense()}")
print(f"\noutput of sklearn implemented TFIDF::\n{skl_output.todense()}")
```
